Features of face_recognition library:

1.Find Faces in Picture.
To find faces in a given picture, we can use the face_locations(image file) function from the Face_recognition library.
2. Comparing Faces :
For comparing faces, we have a method called compare_faces() in the facial recognition library.
compare_faces() returns a list of True/False values indicating which known_face_encodings match the face encoding to check

3.And also we can see the face_encodings() method, which returns the 128-dimension face encoding for each face in the image.

```python
from google.colab.patches import cv2_imshow

def read_img(path):
    img = cv2.imread(path)
    (h, w) = img.shape[:2]
    width = 500
    ratio = width / float(w)
    height = int(h * ratio)
    return cv2.resize(img, (width, height))


known_encodings = []
known_names = []
known_dir = 'mini_proj-dataset/Abhishek'

for file in os.listdir(known_dir):
    img = read_img(known_dir + '/' + file)
    img_enc = face_recognition.face_encodings(img)[0]
    print(img_enc)
    known_encodings.append(img_enc)
    known_names.append(file.split('.')[0])
```

```python
# function to convert the JavaScript object into an OpenCV image
def js_to_image(js_reply):
  """
  Params:
          js_reply: JavaScript object containing image from webcam
  Returns:
          img: OpenCV BGR image
  """
  # decode base64 image
  image_bytes = b64decode(js_reply.split(',')[1])
  # convert bytes to numpy array
  jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
  # decode numpy array into OpenCV BGR image
  img = cv2.imdecode(jpg_as_np, flags=1)

  return img
```
